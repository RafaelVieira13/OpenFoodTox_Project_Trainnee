# Prepare The DataFrame

__1. Join the Dataset 'STUDY' with the DataSet 'ENDPOINT' based on the column 'TOX_ID'__

* TOX_ID: Unique identifier links to the ENDPOINTSTUDY table
* Let's Create The following Table: TOX_ID,TRX_ID, STUDY_CATEGORY, TESTTYPE, SPECIES, ROUTE, EXP_DURATION_DAYS, ENDPOINT, QUALIFIER, VALUE, DOSEUNIT, SUB_OP_CLASS, IS_MUTAGENIC, IS_GENOTOXIC, IS_CARCINOGENIC

```python
# Let's take a look to the dataset 'STUDY'
df_study.columns
```

```python
# Let's take a look to the dataset 'ENDPOINT'
df_endpoint.columns
```

```python
# Let's merge both datsets
endpoint_study = pd.merge(df_endpoint[['TOX_ID','STUDY_CATEGORY','TESTTYPE','SPECIES','ROUTE','EXP_DURATION_DAYS','ENDPOINT','QUALIFIER','VALUE','DOSEUNIT']]
                          ,df_study[['TOX_ID','OP_ID','SUB_COM_ID','IS_MUTAGENIC','IS_GENOTOXIC','IS_CARCINOGENIC']],
                          on='TOX_ID', how='left')
endpoint_study
```

__2. Join the Dataset 'STUDY+ENDPOINT' with the DataSet 'component' based on the column 'SUB_COM_ID'__

* Let's get the 'SUB_COM_ID', 'SUB_NAME' and 'SUBPARAM_CODE'

```python
# Let's take a look to the dataset 'COMPONENT'
df_comp.columns
```

```python
endpoint_study_comp = pd.merge(endpoint_study, df_comp[['SUB_COM_ID','SUB_NAME','SUBPARAM_CODE','COM_CASNUMBER','COM_ECSUBINVENTENTRYREF','MOLECULARFORMULA','SMILESNOTATION']],
                              on='SUB_COM_ID',how='left')
endpoint_study_comp
```

```python
endpoint_study_comp.isnull().sum()
```

```python
endpoint_study_comp.shape
```

__3. Join the Dataset 'STUDY+ENDPOINT' with the DataSet 'component' based on the column 'OP_ID'__

* Let's get the 'OP_ID', 'AUTHOR', 'PUBLICATIONYEAR', 'URL'

```python
df = pd.merge(endpoint_study_comp, df_opinion[['OP_ID','AUTHOR','PUBLICATIONYEAR','URL']],
                              on='OP_ID',how='left')
df
```

```python
# Let's delete the ID Columns
df.drop(columns=['TOX_ID','OP_ID','SUB_COM_ID'], inplace=True)
df
```

```python
# Let's order the columns
df = df[['SUB_NAME','SUBPARAM_CODE','COM_CASNUMBER','COM_ECSUBINVENTENTRYREF','MOLECULARFORMULA','SMILESNOTATION','AUTHOR','PUBLICATIONYEAR','STUDY_CATEGORY','TESTTYPE','SPECIES','ROUTE','EXP_DURATION_DAYS',
    'ENDPOINT','QUALIFIER','VALUE','DOSEUNIT','IS_MUTAGENIC','IS_GENOTOXIC','IS_CARCINOGENIC','URL']]

df
```

```python
df.columns
```

```python
# Let's Rename Some Columns
df.rename(columns={'SUB_NAME':'Substance',
                  'SUBPARAM_CODE':'PARAM Code',
                  'COM_CASNUMBER':'CA Number',
                  'COM_ECSUBINVENTENTRYREF':'EC Ref No.',
                  'MOLECULARFORMULA':'Molecular Formula',
                  'SMILESNOTATION':'Smiles',
                  'AUTHOR':'Author',
                  'PUBLICATIONYEAR':'Year',
                  'STUDY_CATEGORY':'Study Category',
                  'TESTTYPE':'Test Type',
                  'SPECIES':'Species',
                  'ROUTE':'Route',
                  'EXP_DURATION_DAYS':'Duration (Days)',
                  'ENDPOINT':'Endpoint',
                  'QUALIFIER':'Qualifier',
                  'VALUE':'Value',
                  'DOSEUNIT':'Unit',
                  'IS_MUTAGENIC':'Mutagenicity',
                  'IS_GENOTOXIC':'Genotoxicity',
                  'IS_CARCINOGENIC':'Carcinogenicity'},inplace=True)
df
```
