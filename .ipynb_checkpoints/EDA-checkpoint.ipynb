__CARCINOGENIC Substances__ 

Number of Positive and Negative

```python
df_study['IS_CARCINOGENIC'].value_counts()
```

```python
# Create the DataFrame
carcinogenic = df_study['IS_CARCINOGENIC'].value_counts().reset_index()
carcinogenic.rename(columns={'index':'carcinogenic_classification', 'IS_CARCINOGENIC':'count'}, inplace=True)
carcinogenic.sort_values(by='count', ascending=False, inplace=True)
carcinogenic
```

```python
# Data To POWERBI
```

```python
plt.figure(figsize=(12,9))
plt.bar(carcinogenic['carcinogenic_classification'],carcinogenic['count'],color='#1f77b4')

plt.title('Carcinogenic Classification', fontweight='bold',fontsize=24)
plt.xlabel('Classification', fontsize=16)
plt.ylabel('Count', fontsize=16)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.show()
```
